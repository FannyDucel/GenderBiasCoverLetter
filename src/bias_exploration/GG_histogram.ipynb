```python
"""Visualization: plot the nb of fields with a Gender Gap of X (histogram)"""

import pylab as pl
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
d = pd.read_csv("../../results/FR/bias_evaluation/sorted_gap+neutre.csv")
#d.gap.hist(bins=20)
sns.set(rc={"figure.figsize":(8, 4)})
sns.set(font_scale=1.5)
plt.axvline(x=0,
            color='black',
            ls='-', 
            lw=2.5)
plt.axvline(x=d.gap.mean(),
            color='red',
            ls='-', 
            lw=2.5)
plt.axvline(x=d.gap.median(),
            color='blue',
            ls='--', 
            lw=2.5)

sns.histplot(d["gap"], element="step")
pl.xlabel("Gender Gap")
pl.ylabel("Number of fields")
plt.legend(loc='upper right', labels=['Ideal mean', 'Mean', 'Median'], fontsize="x-small")
pl.savefig("../../results/FR/bias_evaluation/gendergap_distribution_meanmedian_neutral_fr.png", bbox_inches="tight",dpi=300)
```

```python
"""French version of the figure"""
d = pd.read_csv("../../results/FR/bias_evaluation/sorted_gap+neutre.csv")
#d.gap.hist(bins=20)
sns.set(rc={"figure.figsize":(8, 4)})
sns.set(font_scale=1.5)
sns.histplot(d["gap"], element="step")
plt.axvline(x=d.gap.median(),
            color='blue',
            ls='--', 
            lw=2.5)
plt.axvline(x=d.gap.mean(),
            color='red',
            ls='-', 
            lw=2.5)
pl.xlabel("Écart Genré")
pl.ylabel("Nombre de domaines")
pl.savefig('../../results/FR/bias_evaluation/gender_gap_distribution_fr.png', bbox_inches="tight")
```
